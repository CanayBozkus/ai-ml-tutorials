```python
dummies = pd.get_dummies(df['purpose'], drop_first=True)
df_clean = pd.concat([df.drop('purpose', axis=1), dummies], axis=1)
```

```python
df_clean.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df_clean.drop('not.fully.paid', axis=1)
y = df_clean['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(X_train, y_train)
```

```python
pred = tree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand_tree = RandomForestClassifier()
```

```python
rand_tree.fit(X_train, y_train)
```

```python
pred_rand_tree = rand_tree.predict(X_test)
```

```python
print(confusion_matrix(y_test, pred_rand_tree))
print(classification_report(y_test,pred_rand_tree))
```
